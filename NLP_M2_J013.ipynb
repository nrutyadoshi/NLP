Preprocessing

```
import nltk
nltk.download("all")
```

```
import re
```

```
def remove_html_tags(text): # removing all the html data 
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```
news["Body"] = news.apply(lambda row: remove_html_tags(row.Body),axis=1) # applying the function on body
```

```
def text_data(str): # taking only the text data
  str_new = " ".join(re.findall("[a-zA-Z]+", str))
  return (str_new)
```

```
news["Title"] = news.apply(lambda row: text_data(row.Title),axis=1) # applying the function on title
```

```
news["Body"] = news.apply(lambda row: text_data(row.Body),axis=1) # applying the function on body
```

```
def tokenize(str_new): # tokenization
  tokens = []
  tokens = nltk.word_tokenize(str_new)
  return (tokens)
```

```
news["Title"] = news.apply(lambda row: tokenize(row.Title),axis=1) # applying the function on title
```

```
news["Body"] = news.apply(lambda row: tokenize(row.Body),axis=1) # applying the function on body
```

```
news.head() # here we see the body data is free of tags & the data is tokenized
```

```
from nltk.corpus import stopwords
```

```
def remove(tokens): # removing stopwords & punctuations
  removed = []
  stopword = stopwords.words('german') 
  punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~=+'''
  for w in tokens:
    if (w.lower() not in stopword and w.lower() not in punctuations):
      removed.append(w.lower())
  return (removed)
```

```
news["Title"] = news.apply(lambda row: remove(row.Title),axis=1) # applying the function on title
```

```
news["Body"] = news.apply(lambda row: remove(row.Body),axis=1) # applying the function on body
```

```
news.head()
```

```
def to_string(lemma): # converting the data from list to string format for tfidf
  str1 = ' '.join([elem for elem in lemma]) 
  return (str1)
```

```
news["Title"] = news.apply(lambda row: to_string(row.Title),axis=1) # applying the function on title
```

```
news["Body"] = news.apply(lambda row: to_string(row.Body),axis=1) # applying the function on body
```

```
news.head() # now the data is of string type
```

Tf Idf

```
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```
vectorizer = TfidfVectorizer() 
vectors = vectorizer.fit_transform(news.Title) # doing tfidf on title
feature_names = vectorizer.get_feature_names()
dense = vectors.todense()
denselist = dense.tolist()
df = pd.DataFrame(denselist, columns=feature_names) # making tthe tfidf matrix
```

```
vectorizer1 = TfidfVectorizer()
vectors1 = vectorizer1.fit_transform(news.Body) # doing tfidf on body
feature_names1 = vectorizer1.get_feature_names()
dense1 = vectors1.todense()
denselist1 = dense1.tolist()
df1 = pd.DataFrame(denselist1, columns=feature_names1)
```

Extracting the topics & printing them

```
def print_topics(model, count_vectorizer, n_top_words): # function to print the topics
    words = count_vectorizer.get_feature_names()
    for topic_idx, topics in enumerate(model.components_):
        print("\nTopic #%d:" % topic_idx)
        print(" ".join([words[i]
                        for i in topics.argsort()[:-n_top_words - 1:-1]]))
```

```
# getting topics from title
topic_title = []
words = vectorizer.get_feature_names()
for topic_idx, topics in enumerate(lda.components_):
  topic_title.append([words[i]
                        for i in topics.argsort()[:-10 - 1:-1]])
topic = []
for sents in topic_title:
  topic.append(' '.join([elem for elem in sents]))
topic
```

```
print("Topics found for Title via LDA:")
print_topics(lda, vectorizer, 10)
```

```
# getting topics from body
topic_body = []
words1 = vectorizer1.get_feature_names()
for topic_idx, topics in enumerate(lda.components_):
  topic_body.append([words1[i]
                        for i in topics.argsort()[:-10 - 1:-1]])
topic1 = []
for sents in topic_body:
  topic1.append(' '.join([elem for elem in sents]))
topic1
```

```
print("Topics found for Body via LDA:")
print_topics(lda1, vectorizer1, 10)
```
