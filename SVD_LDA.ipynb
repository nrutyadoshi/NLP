```
import pandas as pd
import numpy as np
```

```
clothes = pd.read_csv("Womens Clothing E-Commerce Reviews.csv")
```

```
clothes.drop("Unnamed: 0",1,inplace=True)
```

```
import nltk
nltk.download("all")
```

```
import re
```

```
def text_data(str):
  str_new = " ".join(re.findall("[a-zA-Z]+", str))
  return (str_new)
```

```
data["Text"] = data.apply(lambda row: text_data(row.Text),axis=1)
```

```
def tokenize(str_new):
  tokens = []
  tokens = nltk.word_tokenize(str_new)
  return (tokens)
```

```
data["Text"] = data.apply(lambda row: tokenize(row.Text),axis=1)
```

```
from nltk.corpus import stopwords
```

```
def remove(tokens):
  removed = []
  stopword = stopwords.words('english') 
  punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
  for w in tokens:
    if (w.lower() not in stopword and w.lower() not in punctuations):
      removed.append(w.lower())
  return (removed)
```

```
data["Text"] = data.apply(lambda row: remove(row.Text),axis=1)
```

```
def tagging(removed):
  pos = []
  pos.append(nltk.pos_tag(removed))
  return (pos)
```

```
data["Text"] = data.apply(lambda row: tagging(row.Text),axis=1)
```

```
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
```

```
def lemmatization(pos):
  lem = WordNetLemmatizer()
  lemma = []
  for word in pos:
    for w in word:
      pos_value = ""
      if (w[1].startswith('J')):
        pos_value = wordnet.ADJ
      elif (w[1].startswith('V')):
        pos_value = wordnet.VERB
      elif (w[1].startswith('N')):
        pos_value = wordnet.NOUN
      elif (w[1].startswith('R')):
        pos_value = wordnet.ADV 
      else:
        continue
      lemma.append(lem.lemmatize(w[0],pos_value))
  return (lemma)
```

```
data["Text"] = data.apply(lambda row: lemmatization(row.Text),axis=1)
```

```
data.head()
```

```
def to_string(lemma):
  str1 = ' '.join([elem for elem in lemma]) 
  return (str1)
```

```
data["Text"] = data.apply(lambda row: to_string(row.Text),axis=1)
```

```
data.head()
```

```
from sklearn.feature_extraction.text import TfidfVectorizer 
```

```
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(data.Text)
feature_names = vectorizer.get_feature_names()
dense = vectors.todense()
denselist = dense.tolist()
df = pd.DataFrame(denselist, columns=feature_names)
```

```
df.head()
```

```
data = pd.concat([data,df],1)
```
