```
import pandas as pd
import numpy as np
```

```
data = pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz",lines=True)
```

```
data = data.iloc[:50]
data.shape
```

```
import nltk
nltk.download('all')
```

```
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```
def pre_process(str):
  words  = nltk.word_tokenize(str)
  words = [word.lower() for word in words]
  stopword = []
  stopword = stopwords.words('english')
  filtered_sentence = [w for w in words if not w in stopword] 
  punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
  filtered_sentence2 = [w for w in filtered_sentence if not w in punctuations]
  unique = []
  for w in filtered_sentence2:
    if w not in unique:
      unique.append(w)
  return (unique)
```

```
data["summary_unique"] = data.apply(lambda row: pre_process(row.summary),axis=1)
```

```
data.head()
```

```
unique = []
for row in data["summary_unique"]:
  for word in row:
    unique.append(word)
```

```
unique
```

```
unique_list = []
for w in unique:
    if w not in unique_list:
      unique_list.append(w)
```

```
unique_list
```

```
import gensim 
from gensim.models import Word2Vec 
```

```
model = Word2Vec([unique_list],min_count=1)
```

```
model.wv.vocab
```

```
for i in unique_list[0:10]:
  for j in unique_list[1:10]:
    if(i<j):
      print(i,j,model.wv.similarity(w1=i,w2=j))
```
