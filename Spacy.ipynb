```
data = data.iloc[:50]
data.shape
```

```
def dictionary_spacy(str):
  dictionary= dict()
  sp = spacy.load("en_core_web_sm")
  doc = sp(str)
  dictionary["tokenize"] = [token.text for token in doc]
  dictionary["lemmatization"] = [token.lemma_ for token in doc]
  dictionary["pos"] = [(token, token.pos_) for token in doc]
  dictionary["dependancy"] = [(token, token.dep_) for token in doc]
  return dictionary
```

```
def noun(str):
  sp = spacy.load("en_core_web_sm")
  doc = sp(str)
  pos = [(token, token.pos_) for token in doc]
  noun = []
  for i in range(0,len(pos)):
    if ("PRO" in pos[i][1]):
      noun.append(pos[i][0])
    elif("NOUN" in pos[i][1]):
      noun.append(pos[i][0])
  return noun
```

```
data["reviewText_spacy"] = data.apply(lambda row: dictionary_spacy(row.reviewText),axis=1)
data["summary_spacy"] = data.apply(lambda row: dictionary_spacy(row.summary),axis=1)
```

```
data["reviewText_noun"] = data.apply(lambda row: noun(row.reviewText),axis=1)
data["summary_noun"] = data.apply(lambda row: noun(row.summary),axis=1)
```
